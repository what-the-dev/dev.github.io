# Plotting the data

The `plot` method on Series and DataFrame is just a simple wrapper around `plt.plot()`

**1. Time Series** <br>
Matplotlib has some annoyances when plotting dates. You can read about that [here](https://matplotlib.org/gallery/recipes/common_date_problems.html). The easiest way to work around these with Pandas is to convert your dates to a datetime objuect using [pd.to_datetime()](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.to_datetime.html) then **set your date as the index**. If the index consists of dates, pandas calls `gcf().autofmt_xdate()` and tries to format the x-axis dates nicely.

```python
# convert date to a datetime object

df.date = pd.to_datetime(df.date)

# set the index to the date
df.index = df.date


df.head()
```

As you can see, the index is now a `datetime` index. This is useful when plotting timeseries data and will make your life much much simpler.</br>

Lets plot some data!

# Line Plot

```python
# plot a single variable as a time series
# to adjust the size of the figure, use figsize as a keyword argument

df['high'].plot.line(figsize=(12,8));
```

```python
# plot mulitple variables as a time series
# use two sets of square brackets to return a data frame [[ ]]

df[['change', 'changeOverTime']].plot.line(figsize=(12,8));
```

```python
# plot multiple variables on separate axis

df[['change', 'changeOverTime']].plot.line(subplots=True, figsize=(12,8));
```

## Customising your plots
- Adding titles
    - Axis titles can be set in a number of ways
    - e.g. like [this](https://stackoverflow.com/questions/21487329/add-x-and-y-labels-to-a-pandas-plot/21487560) or [this](http://jonathansoma.com/lede/data-studio/matplotlib/labeling-your-axes-in-pandas-and-matplotlib/)
- Setting colours
- adjusting line types
- setting axis limits

```python
# there are a few ways to set axis labels and titles
# when you create a plot, it returns an ax element
# there are a tonne of methods available to use
# like 'set_xlabel' and 'set_ylabel'
# or you can set both using 'set'

# set an axis object
ax = df[['change', 'changePercent']].plot.line(figsize=(12,8))

# Set the title
ax.set_title("My First Plot with Pandas")

# Set the y label
ax.set_ylabel("Ch-Ch-Changes")

# Set the x label
ax.set_xlabel("Date");
```

```python
# here is another way to do the same thing
# using 'set'

# set an axis object
ax = df[['change', 'changePercent']].plot.line(figsize=(12,8))

ax.set(title="My Second Plot with Pandas", 
      ylabel='Ch-Ch-Changes', xlabel='Date');
```

```python
# and finally, a third way

df[['change', 'changePercent']].plot.line(figsize=(12,8), title='My Third Plot with Pandas').set(ylabel='Ch-Ch-Changes', xlabel='Date');
```

## Set colours, line type and limit axis
- Colours can be set using the `color` argument. You can supply [hex](https://www.color-hex.com) colour codes, or by using [abbreviations](https://matplotlib.org/2.1.1/api/_as_gen/matplotlib.pyplot.plot.html)
- Line style can be controlled using the `marker` and `linewidth` arguments
- More info on styling can be found [here](https://matplotlib.org/2.1.1/api/_as_gen/matplotlib.pyplot.plot.html) and [here](https://matplotlib.org/tutorials/introductory/customizing.html)
- **axis limits** can be specified using `ax.set()` and `xlim` or `ylim`

```python
# to specify different line and marker options the easiest way is to split out your plot
# so that lines can be controlled individually

fig, ax = plt.subplots(figsize=(12,8))

df.plot(kind='line', y='change', ax=ax, color='#005b96', marker='o')
df.plot(kind='line', y='changePercent', ax=ax, color='r', marker='+', linewidth=2)

ax.set(title="Plot with Marker styles, colours & axis limits", 
       xlabel='Dtae', ylabel='Change', 
       ylim=(-5, 12), 
       xlim=('2018-07-01', '2018-09-01'));
```

# Box Plot

```python
# Box PLot

ax = df[['low', 'high', 'open', 'close']].plot.box( figsize=(12,8), color='#005b96')
ax.set(title='Box-plotting with Pandas');
```

# Scatter Plot

```python
# scatter plot

ax = df.plot.scatter(x='high', y='low', figsize=(12,8), color='#ae0001')
ax.set(title='Scatter-plotting with Pandas');
```

```python
# kde plot

ax = df.plot.kde(y='low', figsize=(12,8))
ax.set(title='Scatter-plotting with Pandas');
```
