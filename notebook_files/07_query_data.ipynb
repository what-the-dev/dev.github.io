# Query Columns with Boolean Expressions

- Using <code>DataFrame.query()</code>
- More info in the [Docs](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.query.html)
- Data can be found [here](file:///Users/devindearaujo/Desktop/bootstrap/06_dropping_data.html)

```python
import pandas as pd
```

```python
df = pd.read_csv('scores.csv', nrows=10)
df
```

### Query using 'equal to'

```python
# single expression
df.query('Team == "Essendon"')
```

### Query using 'NOT equal to'

```python
df.query('Team != "Essendon"')
```

### Query using 'OR' pipe

```python
df.query('Team == "Essendon" | Team == "Fitzroy"')
```

### Query using 'AND' ampersand

```python
df.query('Team == "Essendon" & G == 4')
```

### Query based on Column

```python
df.query('G > B')
```

```python
df.query('F < A')
```
